```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,
                         figsize=(12, 8), sharey='row', sharex=True) # Added 'sharex=True'

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label='$cos x$')
axes[0, 0].plot(x, cos_x**2, color='r',
                linestyle=':', label='$cos^2 x$')
axes[0, 0].set_title('cos $x$ & cos$^2x$') # Moved $ so only 'x's are italicised
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -2, color='r', alpha=.125) # Changed 'g' to 'r', '-1' to '-2'
axes[0, 0].legend(loc='upper right', frameon=True) # Added legend
axes[0, 0].set_xlim(-np.pi, np.pi) # Added x-axis limit
# Removed axes[0, 0].set_xlabel('Angle (radians)')
# Removed axes[0, 0].text(0, 0, 'I am [0, 0]!', fontsize=30, horizontalalignment='center')

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label='$sin x$')
axes[0, 1].set_title('sin $x$ & sin$^2x$') # Added title
axes[0, 1].fill_between(x, sin_x, -2, color='g', alpha=.125) # Changed 'r' to 'g', 'cos_x' to 'sin_x'
axes[0, 1].plot(x, sin_x**2, label='$sin^2 x$')
axes[0, 1].set_ylabel('Cosine Value')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].legend(loc='upper right', frameon=True) # Changed 'lower' to 'upper', 'frameon=False' to 'frameon=True'
# Removed axes[0, 1].text(0, 0, 'I am [0, 1]!', fontsize=30, horizontalalignment='center')

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label='sin $2x$') # Changed label
axes[1, 0].fill_between(x, fun1_x, -25, color='b', alpha=.125) # Changed '0' to '-25'
axes[1, 0].set_title('$e^{-x}$cos $5x$') # Moved $ so only 'x's are italicised
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels(['$-\\pi$', '$-\\pi/2$', '0', '$\\pi/2$', '$\\pi$']) # Added '\\' instead of '\'
axes[1, 0].legend()
# Removed axes[1, 0].text(0, 0, 'I am [1, 0]!', fontsize=30, horizontalalignment='center')

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label='sin $2x$') # Changed label
axes[1, 1].set_title('$e^{-x}$sin $2x$') # Moved $ so only 'x's are italicised
axes[1, 1].fill_between(x, fun2_x, -25, color='y', alpha=.125) # Changed '10' to '-25'
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].set_xticklabels(['$-\\pi$', '$-\\pi/2$', '0', '$\\pi/2$', '$\\pi$']) # Added x-axis labels
axes[1, 1].legend(loc='upper right', frameon=True) # Added 'loc='upper right', frameon=True'
axes[1, 1].set_ylim(-25, 14.9) # Added y-axis limit
axes[1, 1].set_xlabel('Angle (radians)') # Added x-axis label
# Removed axes[1, 1].text(0, 0, 'I am [1, 1]!', fontsize=30, horizontalalignment='center')

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.5)

plt.tight_layout()

plt.savefig('plotting-exercises(good)-plot.jpg', dpi=150) # To save as jpg

plt.show() # If this line is placed before plt.savefig(), plot only shows up in notebook but not jpg

# Removed all '\'

# How to remove empty white spaces on right and left side
```
