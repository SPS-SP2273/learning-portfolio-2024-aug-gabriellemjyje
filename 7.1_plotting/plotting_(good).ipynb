# Multiple rows and columns

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=3,
                       figsize=(10, 3),
                       sharex=True, 
                       sharey=True)
for a in ax.flatten():
    a.plot(x, sin_x)

plt.tight_layout()
```

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=3,
                       figsize=(10, 3),
                       sharex=True, 
                       sharey=True)
for count, a in enumerate(ax.flatten()):
    if count == 3:
        a.plot(x, 10*cos_x, 'r')
        continue
a.plot(x, sin_x)
a.grid(alpha=.25)
a.spine['right'].set_visible(False) # Despining
a.spine['top'].set_visible(False)

ax[-1, -1].set_xlim(-10,10)

plt.tight_layout()
```
