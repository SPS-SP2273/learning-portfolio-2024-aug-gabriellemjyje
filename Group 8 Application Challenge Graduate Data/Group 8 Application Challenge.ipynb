# Task 4

## Shorter names

```python
df_employment['university'].unique()
```

```python
# Creating a dictionary for abbreviated university names
replace_info = {'Nanyang Technological University': 'NTU',
       'National University of Singapore': 'NUS',
       'Singapore Management University': 'SMU',
       'Singapore Institute of Technology': 'SIT',
       'Singapore University of Technology and Design': 'SUTD',
       'Singapore University of Social Sciences': 'SUSS'}

# Replacing university names with abbreviated names specified in the dictionary
df_employment.replace(to_replace=replace_info, inplace=True)
df_employment
```

## Replace na with np.nan

```python
df_employment.replace('na', np.nan, inplace=True)
df_employment
```

## Change type to float

```python
new_type_info = {'basic_monthly_median': 'float'}
df_employment = df_employment.astype(new_type_info)
df_employment.info()
```

# Task 5

```python
table = df_employment.groupby(by=['year', 'university'])[['basic_monthly_median']].agg(['count', 'min', 'max', 'mean', 'std']).round(3) # Creating table
# Grouping data into year and university (rows)
# Outermost square brackets outside 'basic monthly median' is for creating a list of headings
# .round(3) rounds off the numbers to 3d.p.
table
```

# Task 6

```python
from matplotlib import pyplot as plt

pt_table = pd.pivot_table(table['basic_monthly_median'], values = 'mean', index = 'year', columns = 'university')
# Explain how pivot table works?
# If take out outer sq brackets in task 5, python can guess that we need to call basic monthly median bc there's only one column
# If put outer sq brackets, there could be more than 1 item in the list (?) so python can't guess that we need only the item we put inside the list
pt_table.plot(kind='bar', figsize=(10, 4))

plt.xlabel('year')
plt.ylabel('Median Salary')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.02, 0.5), loc='center left') 
# Adjusting position of legend with respect to specified coordinates
# (1.02, 0.5) is center left of the box

plt.tight_layout()
plt.show()
```
